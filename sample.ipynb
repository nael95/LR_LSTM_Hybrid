```python
import pandas as pd
import numpy as np

#import sample dataset (CIMB daily stock for 10 days)
sample = pd.read_csv("sample cimb.csv")

#display to check if dataset is correct
sample
```

```python
#change entries in Date column to date datatype
sample['Date'] = pd.to_datetime(sample['Date'])

sample.set_index('Date', inplace=True)
sample_datea = sample[['Price', 'Open', 'High', 'Low']]
```

```python
import numpy as np
def create_sequences(sample_datea, seq_length=3):
    X, y = [], []
    for i in range(len(sample_datea) - seq_length):
        X.append(sample_datea[i:i+seq_length])
        y.append(sample_datea[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 3
X, y = create_sequences(sample_datea['Price'].values, seq_length)
X = X.reshape((X.shape[0], X.shape[1], 1))  # Add feature dimension
```

```python
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

## Implementing Lag to a copy of the sample dataset

```python
# Copy the dataframe to avoid modifying the original
data = sample_datea.copy()

for i in range(1, 6):  # 1 to 10
    data[f'Lag_{i}'] = data['Price'].shift(i)

# Drop NaN rows caused by shifting
data.dropna(inplace=True)

print(data.head())  # Check if lags were created correctly

# Drop NaN values caused by shifting
data.dropna(inplace=True)

```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression

# ✅ Step 1: Scale 'Price' before using it as y_lin
scaler_price = MinMaxScaler()
data['Price_scaled'] = scaler_price.fit_transform(data[['Price']])  # Scale only 'Price'

# ✅ Step 2: Define features (X) and target variable (y)
X_lin = data[['Lag_1', 'Lag_2', 'Lag_3', 'Lag_4', 'Lag_5']]  # Input features
y_lin = data['Price_scaled']  # Use the **scaled Price** as the target variable
data = data.dropna().reset_index(drop=True)

# ✅ Step 3: Train-test split
train_size = int(len(X_lin) * 0.8)
X_train_lin, X_test_lin = X_lin[:train_size], X_lin[train_size:]
y_train_lin, y_test_lin = y_lin[:train_size], y_lin[train_size:]

# ✅ Step 4: Train Linear Regression model
lin_model = LinearRegression()
lin_model.fit(X_train_lin, y_train_lin)

# Print model coefficients
print("Model Coefficients:", lin_model.coef_)
print("Model Intercept:", lin_model.intercept_)

#  Step 5: LSTM Predictions
X_test_lstm = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
lstm_predictions = lstm_model.predict(X_test_lstm)

#  Fix inverse transform issue
lstm_predictions = scaler_price.inverse_transform(lstm_predictions)  # Use scaler_price!
```

### Checking if Model is Legit

### Assessing Model using Metrics (RMSE, MAE and R-squared)
